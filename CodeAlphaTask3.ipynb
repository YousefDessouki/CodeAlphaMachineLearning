```python
from tqdm import tqdm
from urllib.request import urlopen
from io import BytesIO
from zipfile import ZipFile

def download_and_unzip(url, extract_to='Datasets', chunk_size=1024*1024):
    http_response = urlopen(url)

    data = b''
    iterations = http_response.length // chunk_size + 1
    for _ in tqdm(range(iterations)):
        data += http_response.read(chunk_size)

    zipfile = ZipFile(BytesIO(data))
    zipfile.extractall(path=extract_to)

# Example usage
download_and_unzip('https://git.io/J0fjL', extract_to='Datasets')
```

Extract the image from the tgz format

```python
import tarfile
import os

# Define the path to the downloaded tgz file
tgz_file_path = '/content/Datasets/IAM_Words/words.tgz'

# Define the directory where you want to extract the images
extracted_images_dir = '/content/Datasets/IAM_Words'

# Create the directory if it doesn't exist
if not os.path.exists(extracted_images_dir):
    os.makedirs(extracted_images_dir)

# Extract the tgz file
with tarfile.open(tgz_file_path, 'r:gz') as tar:
    tar.extractall(extracted_images_dir)

print("Images extracted successfully!")
```

Extract all text for each image in words.txt file

```python
with open("/content/Datasets/IAM_Words/words.txt", "r") as file:
    lines = file.readlines()

extracted_words = []
first_file_path=[]
second_file_path=[]
third_file_path=[]

for line in lines:
    # Skip lines starting with '#' or containing 'err'
    if line.startswith('#') or 'err' in line:
        continue

    # Split the line by whitespace
    parts = line.split()

    # Extract the last word
    last_word = parts[-1]

    filename=parts[0]
    # Split the filename into two parts
    parts = filename.split('-')

    # Extract the first part (a01)
    first_part = parts[0]

    # Join the first two parts to create the second part (a01-000u/)
    second_part = '-'.join(parts[:2]) + '/'

    # Add the last word to the list
    extracted_words.append(last_word)
    first_file_path.append(first_part)
    second_file_path.append(second_part)
    third_file_path.append(filename)

print(extracted_words)
```
